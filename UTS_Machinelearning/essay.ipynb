Jawab :
    Machine learning adalah kemampuan komputer untuk belajar dari data dan membuat keputusan atau prediksi tanpa harus diprogram secara eksplisit. bisa dikatakan machine learning memungkinkan komputer untuk belajar dari pengalaman dan membuat keputusan atau tugas tertentu tanpa harus diatur secara khusus.

2.	Berikan contoh penerapan implemetasi dalam kehidupan sehari-hari terkait Machine Learning. Dan jelaskan mengapa membutuhkannya dan manfaatnya!
Jawab :
    Pengenalan Wajah : Pengenalan wajah di aplikasi seperti penguncian layar ponsel atau penandaan teman di media sosial menggunakan machine learning. Sistem ini belajar dari fitur wajah Anda dan dapat mengenali wajah secara akurat. 
    Mengapa Membutuhkan Machine Learning dan Manfaatnya: 
    •	Efisiensi: Otomatisasi tugas-tugas yang repetitif dan kompleks, menghemat waktu dan upaya manusia.
    •	Personalisasi: Pemberian layanan dan rekomendasi yang disesuaikan dengan preferensi dan kebutuhan individual.
    •	Prediksi yang Akurat: Mampu membuat prediksi berdasarkan data historis, seperti cuaca, prediksi penjualan, atau pergerakan pasar.
    •	Deteksi Anomali: Mengidentifikasi pola yang tidak biasa atau mencurigakan dalam data, seperti deteksi fraud atau kegagalan peralatan.
    •	Peningkatan Keamanan: Penggunaan machine learning dalam sistem keamanan untuk mendeteksi dan menanggapi ancaman dengan cepat.

3. Jelaskan macam-macam taxonomi dalam pengerapan Machine Learning!
Jawab :
    Berdasarkan Jenis Pembelajaran:
    a.  Supervised Learning (Pembelajaran Terawasi):
        -	Klasifikasi: Memprediksi kelas atau label dari input tertentu.
        -	Regresi: Memprediksi nilai kontinu berdasarkan input.
    b.	Unsupervised Learning (Pembelajaran Tanpa Pengawasan):
        -	Clustering: Mengelompokkan data ke dalam kelompok-kelompok yang serupa.
        -	Reduksi Dimensi: Mengurangi jumlah fitur atau variabel dalam data.
    c.	Reinforcement Learning (Pembelajaran Penguatan):
        -	Agent belajar melalui interaksi dengan lingkungan untuk memaksimalkan hadiah atau nilai tertentu.

    Berdasarkan Tujuan atau Tugas:
    -	Klasifikasi: Memasukkan data ke dalam kategori atau kelas tertentu.
    -	Regresi: Memprediksi nilai numerik atau kontinu.
    -	Clustering: Mengelompokkan data yang memiliki kesamaan tertentu.
    -	Reduksi Dimensi: Mengurangi jumlah fitur atau variabel dalam data.

    Berdasarkan Jenis Algoritma:
    -	Decision Tree: Model berstruktur pohon keputusan.
    -	Random Forest: Kombinasi dari beberapa pohon keputusan.
    -	K-Nearest Neighbors (KNN): Mengklasifikasikan berdasarkan mayoritas kelas tetangga terdekat.
    -	Neural Networks (Deep Learning): Jaringan saraf dengan lapisan-lapisan yang mendalam.
    -	Support Vector Machines (SVM): Mencari batas keputusan optimal.
    -	K-Means: Algoritma clustering berbasis pusat kelompok.

    Berdasarkan Jenis Data:
    -	Structured Data: Data terorganisir dalam format tabel atau database.
    -	Unstructured Data: Data tanpa format atau struktur yang jelas, seperti teks, gambar, atau suara.

    Berdasarkan Skala Implementasi:
    -	Batch Learning: Model dilatih menggunakan seluruh dataset sekaligus.
    -	Online Learning: Model diperbarui secara kontinu saat menerima data baru.

    Berdasarkan Interaksi Manusia:
    -	Human-in-the-Loop (HITL): Kombinasi intervensi manusia dengan proses machine learning.
    -	Machine Learning Ops (MLOps): Otomatisasi dan manajemen siklus hidup model secara end-to-end.
